```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interactive, interact
```

```python
predictions = np.load('../outputs/detection_focal_preds.npz')
ids, masks, labels = predictions.values()
ids.shape, masks.shape,  labels.shape
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
@interact
def plot_mask(name=ids):
    img = np.load(f'../data/testing/slices/sagittal/{name}.npy')
    fig, ax = plt.subplots(1,1, figsize=(10, 10))
    ax.axis('off')
    n = list(ids).index(name)
    mask = sigmoid(masks[n, 0])
    mask = np.where(mask < 0.25, np.nan, mask)
    ax.imshow(img)
    ax.imshow(mask, alpha=0.7, cmap='winter')
```
